```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
schools_complete = "Resources/schools_complete.csv"
students_complete = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools_data = pd.read_csv(schools_complete)
students_data = pd.read_csv(students_complete)

# Combine the data into a single dataset.  
education_df = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])

education_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(education_df["School ID"].unique())
total_students = len(education_df["Student ID"].unique())

total_budget = education_df["budget"].unique().sum()
avg_math = education_df["math_score"].mean()
avg_reading = education_df["reading_score"].mean()


math_pass_df = education_df.loc[education_df["math_score"] >= 70]
num_math_pass = len(math_pass_df)
percent_math_pass = (num_math_pass/total_students)*100

read_pass_df = education_df.loc[education_df["reading_score"] >=70]
num_read_pass = len(read_pass_df)
percent_read_pass = (num_read_pass/total_students)*100

both_pass_df = education_df.loc[(education_df["math_score"] >=70) & (education_df["reading_score"] >=70)]
num_both_pass = len(both_pass_df)
percent_both_pass = (num_both_pass/total_students)*100

district_summary_df = pd.DataFrame(
    data={
        "Total Schools": total_schools,
        "Total Students": total_students,
        "Total Budget": total_budget,
        "Average Math Score": avg_math,
        "Average Reading Score": avg_reading,
        "% Passing Math": percent_math_pass,
        "% Passing Reading": percent_read_pass,
        "% Overall Passing": percent_both_pass
    },
    index=[0]
)

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
schools = list(education_df["school_name"].unique())
i = 0
school_summary_df = pd.DataFrame()

for school in schools:
    school_df = education_df.loc[education_df["school_name"] == school]
    school_name = list(school_df["school_name"].unique())[0]
    school_type = list(school_df["type"].unique())[0]
    students = len(school_df["Student ID"])
    budget = list(school_df["budget"])[0]
    stud_budget = budget/students
    m_score = school_df["math_score"].mean()
    r_score = school_df["reading_score"].mean()

    m_pass_df = school_df.loc[school_df["math_score"] >= 70]
    num_m_pass = len(m_pass_df)
    percent_m_pass = (num_m_pass/students)*100

    r_pass_df = school_df.loc[school_df["reading_score"] >=70]
    num_r_pass = len(r_pass_df)
    percent_r_pass = (num_r_pass/students)*100

    pass_df = school_df.loc[(school_df["math_score"] >=70) & (school_df["reading_score"] >=70)]
    num_pass = len(pass_df)
    percent_pass = (num_pass/students)*100
    
    size = list(school_df["size"].unique())[0]

    temp_df = pd.DataFrame(
        data={
            "School": school_name,
            "School Type": school_type,
            "Total Students": students,
            "Budget": budget,
            "Budget per Student": stud_budget,
            "Average Math Score": m_score,
            "Average Reading Score": r_score,
            "% Passing Math": percent_m_pass,
            "% Passing Reading": percent_r_pass,
            "% Overall Passing": percent_pass,
            "Size": size
        },
        index=[i]
    )
    i += 1
    school_summary_df = school_summary_df.append(temp_df)

school_summary_df






```

  * Average Reading Score

  * Overall Passing Rate (Average of the above two)

```python
b_one_df = school_summary_df.loc[school_summary_df["Budget per Student"] <= 584]
b_two_df = school_summary_df.loc[(school_summary_df["Budget per Student"] >= 585) & (school_summary_df["Budget per Student"] <= 629)]
b_three_df = school_summary_df.loc[(school_summary_df["Budget per Student"] >= 630) & (school_summary_df["Budget per Student"] <= 644)]
b_four_df = school_summary_df.loc[school_summary_df["Budget per Student"] >= 645]

df_list = [b_one_df, b_two_df, b_three_df, b_four_df]
df_names = ["< $584", "\$585-$629", "\$630-$644", ">$645"]
spb_df = pd.DataFrame()
i = 0

for df in df_list:
    name = df_names[i]
    a_math = df["Average Math Score"].mean()
    a_read = df["Average Reading Score"].mean()
    p_math = df["% Passing Math"].mean()
    p_read = df["% Passing Reading"].mean()
    p_overall = df["% Overall Passing"].mean()
    
    
    temp_df = pd.DataFrame(
        data={
            "Budget Range": name,
            "Average Math Score": a_math,
            "Average Reading Score": a_read,
            "Percent Passing Math": p_math,
            "Percent Passing Reading": p_read,
            "Overall Passing Rate": p_overall,
        },
        index=[i]
    )
    i += 1
    spb_df = spb_df.append(temp_df)

spb_df
```
